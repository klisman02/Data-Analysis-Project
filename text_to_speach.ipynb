```python
from gtts import gTTS

text_to_say = "How are you doing?"

language = "en"

gtts_object = gTTS(text=text_to_say, lang=language, slow=False)

gtts_object.save("output.mp3")
```

```python
from IPython.display import Audio

Audio("output.mp3")
```

```python
french_text = "Bonjour, comment allez-vous?"

french_language = "fr"

french_gtts_object = gTTS(text=french_text, lang=french_language, slow=False)

french_gtts_object.save("french_output.mp3")

Audio("french_output.mp3")
```
