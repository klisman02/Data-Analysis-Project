```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
import requests
url = 'https://s5.static.brasilescola.uol.com.br/be/2022/01/panthera-tigris.jpg'
r = requests.get(url)

# Load the image directly from the content using cv.imdecode
img_array = np.asarray(bytearray(r.content), dtype=np.uint8)
img = cv.imdecode(img_array, cv.IMREAD_COLOR)
```

```python
# Check if the image was loaded successfully
if img is not None:
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    # Corrected function name for Gaussian blur
    img = cv.GaussianBlur(gray, (3,3), 0)
else:
    print("Error: Could not load image.")
```

```python
sobelx = cv.Sobel(src=img, ddepth=cv.CV_64F, dx=1, dy=0, ksize=5)

sobely = cv.Sobel(src=img, ddepth=cv.CV_64F, dx=0, dy=1, ksize=5)

sobelxy = cv.Sobel(src=img, ddepth=cv.CV_64F, dx=1, dy=1, ksize=5)

plt.figure(figsize=(18,19))
plt.subplot(221)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(sobelxy, cmap = 'gray')
plt.title('Sobel X Y')
plt.axis('off')

plt.subplot(223)
plt.imshow(sobelx, cmap = 'gray')
plt.title('Sobel X')
```
